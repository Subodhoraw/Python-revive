```python
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [45000, 52000, 48000, 61000, 58000, 67000]

plt.plot(months, sales, color='purple', linewidth=2, linestyle='--', marker='s')
plt.grid(True)
plt.title('Monthly Sales Growth', fontsize=16, color='navy')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.show()
```

```python
""" Bar chat """
products = ['product A', 'product B' , 'Product C','product D']
revenue = [35000, 48000, 29000,55000]
plt.bar(products,revenue , color = 'skyblue')
plt.title('Revenue by products')
plt.xlabel('product')
plt.ylabel('Revenue ($)')
plt.show()
```

```python
"""histogram"""
ages = [23,28,25,22,26,30,24,38,27,26,28,31,39]
plt.hist(ages, bins=5,color = 'lightgreen',edgecolor ='black')
plt.title('Age Distribution')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
""" scatter Plot"""
study_hours = [1,2,3,4,5,6,7,8,9]
test_scores =[45,55,60,65,70,85,88,95, 70]
plt.scatter(study_hours,test_scores,color ='red')
plt.title('study hours vs Test score ')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.show()
```

```python
"""pie chart"""
categories = ['marketing','Operations','Sales','IT']
expenses = [25000, 35000,3000,20000]

plt.pie(expenses, labels = categories, autopct ='%1.1f%%')
plt.title('Department Expenses')
plt.show()
              
              
```
