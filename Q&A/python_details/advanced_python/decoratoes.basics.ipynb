## decorator??
### a decorator is a function thst modifies another function without editing its code 
✔ Adding extra features
✔ Without touching the main logic
✔ Just like adding a case to a phone (functions stays same, features increase)

```python
def my_decorator(func):
    def wrapper():
        print(" before opening says ")
        func()
        print("after opening")
    return wrapper

@my_decorator
def hello():
    print("hello")

hello()


```
