### clouser 
clouser = a function remembering values from its outer function even after the outer functon is finished 
Think of closure like:

✔ A function storing some memory
✔ Used for data hiding
✔ Used for encapsulation
✔ Used in decorators
✔ Used in real-world apps (payment, authentication, caching, logging)


```python
def wallet(pin):
    def check_balance(amount):
        print(f" Access allowed with {pin}")
        print(f" Your balance is: {amount}")
    return check_balance
mywallet = wallet(1234)
mywallet(5000)
mywallet(90000)

```

```python
def payment_gateway(fee):
    def process(amount):
        net = amount - (amount* fee)
        return net
    return process
razorpay = payment_gateway(0.02)
stripe = payment_gateway(0.03)
print(razorpay(980))
```
