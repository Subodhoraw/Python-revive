```python
import pandas as pd
df = pd.DataFrame({
    'Date': ['2024-01','2024-01', '2024-02','2024-02'],
    'Product':['A', 'B','C','D'],
    'Sales': [150, 160,170,180],
    'Region':[ 'East','West','East','West']
})
print(df)

#create pivot table 
pivot = df.pivot_table(
    values = 'Sales',
    index = 'Date',
    columns ='Product',
    aggfunc='sum')
print(f"\n{pivot}")
```
