```python
#1.2 DataFrame(2d data
# A data frame is like excel sheeet rows and columns
#Method 1 : from dictionary
data = {
    'Name': ['Alice','Bob','Charlie','David'],
    'Age':[25,30,35,28],
    'city':['NYC','LA','Chicago','Boston'],
    'Salary': [70000,80000,75000,72000]
}
df=pd.DataFrame(data)
print(df)

#method 2 from list
data =[ ['Alice',25,'NYC'],
['Bob',30,'LA'],
['Charlie',34,'Chicago']]
df =pd.DataFrame(data,columns=['Name','Age','City'])
print(df)

#Method 3 : From Numpy Array
arr = np.array([[1,2,3],[4,5,6]])
df = pd.DataFrame(arr, columns = ['A','B','c'])
print(df)

#Method 4: from  list of dictionaries
data = [
    {'Name': 'Alice','age':30},
    {'Name': 'Bob','Age':30}
]
df =pd.DataFrame(data)
print(df)

```

# basics operations

```python
#create sample Dataframe 
import pandas as pd 
df =pd.DataFrame({
    'Name':['Alice','Bob','Charlie','Subodh','David','Eve'],
    'Age': [ 25,34,50,24,23,30],
    'City':['NYC','LA','Chicago','Kolkata','HyD','HTC'],
    'Salary':[70000,80000,90000,100000,110000,89000]})
df.head() # it will print all rows
df.head(3) #it will print only 3 rows 
df.tail() # it will print last 5 rows 
df.tail(3) # it will print last 2 rows 

#Data frame info
df.shape #(6,4) - 6 rows 5 columns 
df.columns # it will provide columns name 
df.index #it will provide row indices
df.dtypes #it will provide data types on of each column 
df.describe() #statistical summary
df.info()# detailed info





```

```python
#Quick Stats
df.describe()
df['Age'].mean()
df['Age'].sum()
df['Age'].min()
df['Age'].max()
df['Age'].std()
```

## Selecting data



```python
df['Name'] #select single columns(returns DataFrame)
df.Name

# select multiple columns
df[['Name','Age']]

# select rows by position
df.iloc[0]
df.iloc[0:3] #it will print 0 to 2 rows
df.iloc[[0,2,4]] # this way we can select rows

#select rows by position 
df.loc[0]
df.loc[0:2]

#select speceific cells
df.iloc[0,1] # it will oprint row and column
df.loc[0, 'Age']

```

```python
df.loc[0,'Age']
```

### select with conditions

```python
df[df['Age'] > 25] #rows where age is more than 25
df[df['City']=='NYC'] #rows where city is the NYC
df[(df['Age']>25) & (df['Salary'] > 75000)] #multiple condition 
```

### adding and modifying data

```python
## add new column
df['Bonus'] = 5000
df['Total'] = df['Salary'] + df['Bonus']

# add seniors
df['Senior'] = df['Age'] >30

#modifying existing column
df['Salary'] = df['Salary'] * 1.1 #10% raise 

# add new rows 
new_row = {'Name': 'Frank','Age': 29, 'City': 'Miami', 'Salary': 73000}
df =pd.concat([df,pd.DataFrame([new_row])], ignore_index =True)
print(df)

# delete row
df = df.drop(0 ,axis = 0)
df = df.drop(index =[0,1])

```
