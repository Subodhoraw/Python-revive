## Quick python and numpy review 

```python
list = [1,2,3,4,5]
list.append(6)
list[0] = 10
print(list)
```

```python
# dictionaries
dict = {'name': 'John','age':30, 'city': 'Boston'}
dict['age'] = 35
print(dict)
```

```python
#list comprehensions
squares = [x**2 for x in range(5)]
print(squares)
```

```python
#lambda functions
add = lambda x,y: x+y
results = add(3,5)
print(results)
```

```python
numbers = [1,2,3,4,5,6]
evens = [ x for x in numbers if x%2 == 0]
print(evens)
```
