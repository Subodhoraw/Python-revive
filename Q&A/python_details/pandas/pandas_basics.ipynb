```python
import pandas as pd
import numpy as np
# xheck version
print(pd.__version__)
```

```python
#create Series from list
s =pd.Series([10,20,30,40,50])
print(s)

#create Series with custom index
s = pd.Series([10,20,30], index = ['a','b','c'])
print(s)
#access elements 
print(s['a'])
print(s[0])

#series from dictionary
data = {'alice' : 95,'Bob': 87,'Charlie': 92}
grades = pd.Series(data)
print(grades)
```

```python
#1.2 DataFrame(2d data
# A data frame is like excel sheeet rows and columns
#Method 1 : from dictionary
data = {
    'Name': ['Alice','Bob','Charlie','David'],
    'Age':[25,30,35,28],
    'city':['NYC','LA','Chicago','Boston'],
    'Salary': [70000,80000,75000,72000]
}
df=pd.DataFrame(data)
print(df)

#method 2 from list
data =[ ['Alice',25,'NYC'],
['Bob',30,'LA'],
['Charlie',34,'Chicago']]
df =pd.DataFrame(data,columns=['Name','Age','City'])
print(df)

#Method 3 : From Numpy Array
arr = np.array([[1,2,3],[4,5,6]])
df = pd.DataFrame(arr, columns = ['A','B','c'])
print(df)

#Method 4: from  list of dictionaries
data = [
    {'Name': 'Alice','age':30},
    {'Name': 'Bob','Age':30}
]
df =pd.DataFrame(data)
print(df)

```
