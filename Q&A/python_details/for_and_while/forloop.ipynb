Basic Concept
A for loop iterates over a sequence (list, tuple, string, range, etc.) and executes code for each item.

```python
for item in sequence:
#code to excecute 
```

```python
""" loop through a list """
fruits = ["apple", "banana","cherry"]
for fruit in fruits:
  print(fruit)
```

```python
"""loop through a range """
for i in  range(5):
    print(i)
```

```python
"""loop through a string"""
name = "python"

for letter in name:
    print(letter)
```

```python
colors =["red", "green", "blue"]
for index,color in enumerate(colors,start = 0):
    print(f"Index{index}:{color}")
```

Basic Concept
A while loop repeats as long as a condition is true. It checks the condition before each iteration.
Syntax:

```python
while condition:
    #code excecute 
```

```python
count = 1 
while count <= 5:
    print(count)
    count += 1
```

```python
""" user input loop """
password = ""

while password != "secret":
    password = input("enter password")
print("acess granted")
```

```python
numbers = [1,2,3,4,5,6]
total = 0
for num in numbers:
    total += num
print(f"sum: {total}")
```
