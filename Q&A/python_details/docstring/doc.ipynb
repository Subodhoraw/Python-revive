### what is docstring ?
#### a documentation string is   the first statement in a module ,function, class or methode 
Docstrings document code - Use them for functions, classes, and modules
Triple quotes - Always use """..."""
First statement - Must be the first thing after definition
Runtime access - Use __doc__, help(), or inspect module
Different from comments - Comments explain "why", docstrings explain "what"
Follow conventions - Google, NumPy, or Sphinx styles
Include examples - Makes documentation more useful
Public APIs - Always document public functions and classes

```python
def greet(name):
    """
    This function greets the person with the given name.
    """
    return f"Hello, {name}!"

# Access docstring
print(greet.__doc__)
# Output: This function greets the person with the given name.
```

```python
"""class docstring """
class dog:
    """ A simple dog class to represent a dog 
    """
    def ___init__(self,name):
        self.name = name 
    def bark(self):
        return f"{self.name} sayas woof"
print(dog.__doc__)

```
