```python
import seaborn as sns 
import matplotlib.pyplot as plt
# Load sample datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')

# See available datasets
print(sns.get_dataset_names())

# Explore tips dataset
print(tips.head())
print(tips.info())
```

totally basics 

```python
tips = sns.load_dataset('tips')

sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Tips vs Total Bill')
plt.show()

# with hue (color by category)
sns.scatterplot(data = tips,x = 'total_bill',y = 'tip', hue = 'time')
plt.title(' Tips vs Total biilll by time ')

# with size and style 
sns.scatterplot(data = tips, x = 'total bill', y = 'tip', hue ='time',size = 'size', style = 'sex')
plt.title('Complex Scatter Plot')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', 
                hue='time', size='size', style='sex')
plt.title('Complex Scatter Plot')
plt.show()
```

```python
flights = sns.load_dataset('flights')

sns.lineplot(data = flights, x = 'year', y = 'passengers', hue = 'month')
plt.title('passenger Over Time')
plt.legend(bbox_to_anchor = (1.05,1),loc = 'upper left')
plt.tight_layout()
plt.show()
```

# Bar plot

```python
# bar plot with mean values
sns.barplot(data = tips, x = 'day', y = 'total_bill')
plt.title('Average bill by data')
plt.show()
```

```python
## with hue
sns.countplot(data = tips, x = 'day', hue = 'sex')
plt.title('Bills by day and gender')
plt.show
```

## COUNT PLOT 

```python
# Count occurrences
sns.countplot(data=tips, x='day')
plt.title('Number of Bills per Day')
plt.show()

# With hue
sns.countplot(data=tips, x='day', hue='sex')
plt.title('Bills by Day and Gender')
plt.show()
```

### 1.4 Distribution plots
### histogram

```python
sns.histplot(data = tips, x = 'total_bill',kde = True)
plt.title('Distribution of totall bill')
plt.show()

#Multiple variable 
sns.histplot(data=tips, x='total_bill', hue='time', multiple='stack')
plt.title('Stacked Histogram')
```

KDE plot

```python
sns.kdeplot(data = tips, x = 'total_bill')
plt.title('KDE by time')
plt.show()
```
