```python
class Dog:
    def sound(self):
        return "Bark"
    def __str__(self):
        return "Dog"
class cat:
    def sound(self):
        return "meaw"
    def __str__(self):
        return "cat"
class birds:
    def sound(self):
        return "cherp"
    def __str__(self):
        return "birds"

animals = [ Dog(),cat(),birds()]
for animal in animals:
    print(f"{animal} says: {animal.sound()}")
```

```python
class Dog:
    def sound(self):
        return "Bark"
    def __str__(self):
        return "Dog"  # This is a string

class Cat:
    def sound(self):
        return "Meow"
    def __str__(self):
        return "Cat"  # This is a string

class Bird:
    def sound(self):
        return "Chirp"
    def __str__(self):
        return "Bird"  # This is a string

animals = [Dog(), Cat(), Bird()]
for animal in animals:
    print(f"{animal} says: {animal.sound()}")
```
