```python
class Dog:
    def sound(self):
        return "Bark"
    def __str__(self):
        return Dog
class cat:
    def sound(self):
        return "meaw"
    def __str__(self):
        return Dog
class birds:
    def sound(self):
        return "cherp"
    def __str__(self):
        return Dog

animals = [ Dog(),cat(),birds()]
for animal in animals:
    print(f"{animal}  says {animal.sound()}")
```
