How data scintists co work on matplot lib 

```python
fig, ax = plt.subplots(figsize = (10,16))
# 2. talk to ax objects explicitly
ax.plot([1,3,5],[5,6,8],label ="Growth")
ax.set_title('Hello world')
ax.set_xlabel("Quarter")
ax.set_ylabel("REveneu ($)")
ax.plot
ax.grid()

#talk to to the figure objects explicitly
fig.suptitle("company report",fontsize = 16 )
fig.set_facecolor("lightgray")
```

```python
import matplotlib.pyplot as plt

# Create 1 figure containing 2 side-by-side axes (graphs)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# --- Manipulate Graph 1 (Left) ---
ax1.plot([1, 2, 3], [1, 4, 9], color='blue')
ax1.set_title("Linear Growth")
ax1.grid(True) # Grid only on left graph

# --- Manipulate Graph 2 (Right) ---
ax2.plot([1, 2, 3], [10, 5, 1], color='red', linestyle='--')
ax2.set_title("Decline")
ax2.grid(True)
# ax2 has no grid

# --- Manipulate the Whole Canvas ---
fig.suptitle("Comparison Dashboard (Main Title)", fontsize=20, color='purple')
fig.tight_layout() # Auto-adjusts spacing so they don't overlap
plt.savefig('img.svg')
plt.show()
```

## Figure = The entire window/page
## Axes = The actual plot area (can have multiple per figure)

```python
fig, axes = plt.subplots(2,2, figsize = (10,8))

# access individual subplots 
axes[0,0].plot([1,2,3],[1,2,3])
axes[0,1].scatter([1,2,3],[1,2,3])
axes[1,0].bar([1,2,3],[1,2,3])
axes[1,1].hist([1,2,3,1,2,3])


fig.suptitle(" four types of grphs")
plt.tight_layout()
```

## 2.2 Advanced Customization
## Axis Limits

```python
x = np.linspace(0,10,100)
y = np.sin(x)

fig,ax = plt.subplots()
ax.plot(x,y)

# set axis limits 
ax.set_xlim(0,10)
ax.set_ylim(-1.5, 1.5)
```

## Axis Scales

```python
x = np.linspace(1,100,100)
y = x**2

fig,(ax1,ax2) = plt.subplots(1,2, figsize = (12,4))

#liner Scale 
ax1.plot(x,y)
ax1.set_title('linear Scale')

#linear scale

ax2.plot(x,y)
ax2.set_yscale('log')
ax2.set_title('log Scale')

```

# tick customization

```python
x =[1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

fig , ax = plt.subplots()
ax.plot(x,y)

# Set custom tick locations
ax.set_xticks([1,2,3,4,5,6,7])
ax.set_yticks([10,20,30,40,50,60,70])

#set custom tick labels
ax.set_xticklabels(['One', 'Two', 'Three', 'Four', 'Five'])
ax.set_yticklabels(['10K', '20K', '30K', '40K', '50K'])

plt.show()

```
