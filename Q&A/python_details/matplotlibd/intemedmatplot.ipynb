```python
import matplotlib.pyplot as plt

# Create 1 figure containing 2 side-by-side axes (graphs)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# --- Manipulate Graph 1 (Left) ---
ax1.plot([1, 2, 3], [1, 4, 9], color='blue')
ax1.set_title("Linear Growth")
ax1.grid(True) # Grid only on left graph

# --- Manipulate Graph 2 (Right) ---
ax2.plot([1, 2, 3], [10, 5, 1], color='red', linestyle='--')
ax2.set_title("Decline")
ax2.grid(True)
# ax2 has no grid

# --- Manipulate the Whole Canvas ---
fig.suptitle("Comparison Dashboard (Main Title)", fontsize=20, color='purple')
fig.tight_layout() # Auto-adjusts spacing so they don't overlap
plt.savefig('img.svg')
plt.show()
```

# tick customization

```python
x =[1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

fig , ax = plt.subplots()
ax.plot(x,y)

# Set custom tick locations
ax.set_xticks([1,2,3,4,5,6,7])
ax.set_yticks([10,20,30,40,50,60,70])

#set custom tick labels
ax.set_xticklabels(['One', 'Two', 'Three', 'Four', 'Five'])
ax.set_yticklabels(['10K', '20K', '30K', '40K', '50K'])

plt.show()

```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x,y)

#add text
ax.text(5, 0.5,'Peak', fontsize = 14, color = 'red' )

# add annotation with arrow
ax.annotate('Maximum',
            xy = (np.pi/2, 1),
            xytext = (2, 0.7),
            arrowprops = dict(arrowstyle = '->',color = 'red'),
            fontsize = 12)

plt.show()
```

 # color Maps

```python
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.random(100)

fig, ax = plt.subplots()
scatter = ax.scatter(x,y, c= colors, s= sizes, alpha = 0.5, cmap = 'viridis')


#add colorbar
plt.colorbar(scatter)
plt.show()
```

The "Why": In Data Science, Reproducibility is law. If you run this code today and I run it tomorrow, we must see the exact same dots. Without the seed, the random numbers would change every time you ran the script, making it impossible to debug or share results. (Why 42? It's a pop-culture reference to The Hitchhiker's Guide to the Galaxy, but any number works).

This is where the magic happens.

c=colors: "Map the color of each dot to the values in the colors array." High numbers might be yellow, low numbers purple.

s=sizes: "Map the size of each dot to the values in the sizes array."

alpha=0.5: This sets Transparency (0 is invisible, 1 is solid).

DS Tip: Always use alpha in scatter plots! If dots overlap, a solid color hides the density. With transparency, overlapping dots become darker, revealing clusters of data.

cmap='viridis': This stands for Colormap. It tells Python how to translate the numbers into colors. 'Viridis' is the default because it

## Popular colormaps:

### Sequential: 'viridis', 'plasma', 'inferno', 'magma'
###  Diverging: 'RdBu', 'coolwarm', 'seismic'
### Qualitative: 'Set1', 'Set2', 'tab10'

In Data Science, a single number (like an average) is dangerous because it hides the truth.

If I tell you "The average depth of this river is 1 meter," you might try to walk across it. But if I add Error Bars showing it ranges from 0.2m to 5m, you realize you might drown.

Error Bars visualize uncertainty or variability. They tell us: "We think the value is here, but it could reasonably be anywhere in this range."

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
y_error = [0.5, 0.7, 0.3, 0.9, 0.6]

fig, ax = plt.subplots()
ax.errorbar(x, y, yerr=y_error, fmt='o-', capsize=5, capthick=2)
ax.set_title('Plot with Error Bars')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# 1. Setup Data
# ==========================================
data = {
    'Patient_ID': ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9', 'P10'],
    'Age':        [25, 30, 45, 50, 60, 65, 70, 35, 40, 55],
    'Cholesterol':[180, 190, 220, 240, 260, 280, 300, 210, 230, 250],
    'Risk_Score': [1.5, 2.0, 4.5, 6.0, 8.5, 9.2, 9.8, 3.5, 5.0, 7.2], # For Color
    'Glucose':    [80, 85, 140, 160, 200, 220, 250, 95, 110, 180]    # For Size
}

df = pd.DataFrame(data)

# ==========================================
# 2. The Plotting Logic
# ==========================================
fig, ax = plt.subplots(figsize=(10, 6))

# Scaling the size: 
# Raw Glucose is ~100. A 100-pixel dot is small. 
# We multiply by 3 to make the differences more obvious visually.
bubble_sizes = df['Glucose'] * 3 

scatter = ax.scatter(
    df['Age'], 
    df['Cholesterol'], 
    c=df['Risk_Score'],  # Color mapped to Risk
    s=bubble_sizes,      # Size mapped to Glucose
    cmap='inferno',      # Color scheme (Black/Purple to Orange/Yellow)
    alpha=0.6,           # Transparency so we see overlaps
    edgecolors='black'   # Black ring around bubbles makes them pop
)

# ==========================================
# 3. Adding Context
# ==========================================
ax.set_title("Patient Health Analysis: Multi-Variate View", fontsize=14)
ax.set_xlabel("Age (Years)")
ax.set_ylabel("Cholesterol Level")
ax.grid(True, linestyle='--', alpha=0.3)

# Add the Color Bar so we know what "Yellow" means
cbar = plt.colorbar(scatter)
cbar.set_label('Risk Score (0-10)')

# Optional: Label the worst case
# Find the index of the max risk patient
worst_patient = df.loc[df['Risk_Score'].idxmax()]
ax.annotate(f"Highest Risk!\n(Gluc: {worst_patient['Glucose']})", 
            xy=(worst_patient['Age'], worst_patient['Cholesterol']),
            xytext=(worst_patient['Age'] - 10, worst_patient['Cholesterol'] + 20),
            arrowprops=dict(facecolor='red', shrink=0.05))

plt.show()
```
