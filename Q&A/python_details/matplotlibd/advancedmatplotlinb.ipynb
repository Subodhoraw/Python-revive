```python
import matplotlib.pyplot as plt
import numpy as np

# ==========================================
# 1. Generate Dummy Data
# ==========================================
np.random.seed(10)

# Sales: Wide spread, high commissions (High Variance)
sales = np.random.normal(80000, 20000, 100) 

# Admin: Consistent, low spread (Low Variance)
admin = np.random.normal(50000, 5000, 100)

# Tech: High salary, plus some massive outliers (Stock options?)
tech = np.random.normal(90000, 10000, 100)
# Manually adding outliers (The CEO and the intern)
tech = np.append(tech, [150000, 160000, 20000])

data_to_plot = [sales, admin, tech]

# ==========================================
# 2. Creating the Box Plot
# ==========================================
fig, ax = plt.subplots(figsize=(10, 6))

# The 'patch_artist=True' allows us to fill the boxes with color
# 'labels' automatically names the x-axis ticks
bp = ax.boxplot(data_to_plot, 
                patch_artist=True,
                labels=['Sales', 'Admin', 'Tech'],
                vert=True) # vert=False makes it horizontal

# ==========================================
# 3. Customizing the Style (Make it Professional)
# ==========================================
# Change the color of the boxes
colors = ['lightblue', 'lightgreen', 'pink']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

# Change the style of the outliers (The "Fliers")
for flier in bp['fliers']:
    flier.set(marker='o', color='red', alpha=0.5)

ax.set_title('Salary Distribution by Department')
ax.set_ylabel('Annual Salary ($)')
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

## A Violin Plot is essentially a "supercharged" Box Plot.

While a Box Plot shows you summary statistics (Median, Quartiles), it hides the shape of the data. A Violin Plot combines the Box Plot with a Histogram (or Density Plot) on the side.

Why use a Violin Plot?
Imagine you have two classes of students:

Class A: Everyone got a B grade (Normal distribution).

Class B: Half failed (F) and half got A's (Bimodal distribution).

A Box Plot might look identical for both (same median, same range). A Violin Plot will look completely different:

```python
import matplotlib.pyplot as plt
import numpy as np

# ==========================================
# 1. Generate Data with Different Shapes
# ==========================================
np.random.seed(42)

# Normal Distribution (The "Bell Curve")
# e.g., Heights of people
data_normal = np.random.normal(100, 20, 200)

# Bimodal Distribution (Two Peaks)
# e.g., Customer spending (Budget shoppers vs. Luxury shoppers)
data_bimodal = np.concatenate([
    np.random.normal(60, 10, 100),  # Group 1 (Low)
    np.random.normal(140, 10, 100)  # Group 2 (High)
])

data_combined = [data_normal, data_bimodal]

# ==========================================
# 2. Create the Violin Plot
# ==========================================
fig, ax = plt.subplots(figsize=(10, 6))

# showmeans=False, showmedians=True (Show the middle line)
parts = ax.violinplot(data_combined, showmedians=True, showextrema=True)

# ==========================================
# 3. Custom Styling (Making it Readable)
# ==========================================
# Matplotlib returns a dictionary of parts that we can color
for pc in parts['bodies']:
    pc.set_facecolor('#D43F3A')
    pc.set_edgecolor('black')
    pc.set_alpha(0.7)

# Style the lines (Min/Max/Median)
parts['cmedians'].set_color('white') # Make median white so it pops
parts['cmins'].set_color('black')
parts['cmaxes'].set_color('black')
parts['cbars'].set_color('black')

# Context
ax.set_title('Comparison: Normal vs. Bimodal Distribution')
ax.set_xticks([1, 2])
ax.set_xticklabels(['Normal (Class A)', 'Bimodal (Class B)'])
ax.set_ylabel('Score / Value')
ax.grid(True, axis='y', alpha=0.3)

plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 5)]

fig, ax = plt.subplots()
parts = ax.violinplot(data, positions=[1, 2, 3, 4] , showextrema= True )
ax.set_title('Violin Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ==========================================
# 1. Generate Random 2D Data (10x10 Grid)
# ==========================================
# Imagine this is a map of temperature in a 10x10 room
data = np.random.rand(10, 10)

# ==========================================
# 2. Plot using imshow
# ==========================================
fig, ax = plt.subplots()

# imshow = "Image Show"
im = ax.imshow(data, cmap='viridis', interpolation='nearest')

# We have to add the colorbar manually
fig.colorbar(im, ax=ax, label='Temperature')

ax.set_title("Room Temperature Heatmap (Raw Matplotlib)")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ==========================================
# 1. Generate Dummy Medical Data
# ==========================================
# Creating a dataset where some columns are related
np.random.seed(42)
data = pd.DataFrame({
    'Age': np.random.randint(20, 80, 100),
    'Weight': np.random.randint(50, 120, 100),
    'Height': np.random.randint(150, 200, 100),
    'Cholesterol': np.random.randint(150, 300, 100),
    'Risk_Score': np.random.rand(100)
})

# Artificially adding a pattern (Correlation):
# Heavier people usually have higher cholesterol in this dummy data
data['Cholesterol'] += data['Weight'] // 2 
data['Risk_Score'] += data['Age'] / 100

# ==========================================
# 2. Calculate Correlation
# ==========================================
# df.corr() returns a matrix where 1.0 = Perfect Match, -1.0 = Opposite
corr_matrix = data.corr()

# ==========================================
# 3. Create the Heatmap
# ==========================================
fig, ax = plt.subplots(figsize=(8, 6))

sns.heatmap(corr_matrix, 
            annot=True,       # Show the numbers inside the boxes
            fmt=".2f",        # Format to 2 decimal places
            cmap='coolwarm',  # Red = High Correlation, Blue = Low
            linewidths=0.5,   # White lines between boxes
            ax=ax)            # Draw on our specific Matplotlib axis

ax.set_title("Health Metrics Correlation Matrix")
plt.show()
```
