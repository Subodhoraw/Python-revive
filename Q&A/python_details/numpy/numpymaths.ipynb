## universal mathematics functioons 


```python
import numpy as np
arr = np.array([0,30,45,50,60,90])
print(np.sin(np.radians(arr)))
print(np.cos(np.radians(arr)))
print(np.tan(np.radians(arr)))


```

# # exponential and logarithm

```python
arr = np.array([1,2,3,4,5,6,10])
print(np.exp(arr)) ## e = 2.71828 it will provide outcome of e to the power x = 1, 2. 3,4,5
print(np.log(arr)) # Natural log
print(np.log10(arr))
print(np.sqrt(arr))

```

### rounding 

```python
arr = np.array([1.234, 2.567, 3.879])
print(np.round(arr,2)) # it will find 2 didgits from point 
print(np.floor(arr))
print(np.ceil(arr))
```

## absolute value 


```python
arr = np.array([-1,-2,3,-4])
print(np.abs(arr)) # it will find mode value of every number or absolute value 

```

### lets do advanced math 

```python
#broadcasting 
arr = np.array([[2,3,4,],
                [11,12,13]])
print(arr + 10)

## lets add 2d array with 2d array 
row = np.array([2,4,5])
print(arr +row )

## column broadcastes 
col = np.array([[11],[12]])
print(row + col)

arr1 = np.array([1,2,3]).reshape(3,1)
print(arr1)
arr2 = np.array([2,4,5,8,9])
results = arr1 + arr2
print(results)

print()
```

## linear algebra 

```python
## matrix multiplication 
A = np.array([[1,4],[6,7]])
B = np.array([[1,18],[23,9]])
c = np.dot(A,B)
C = A @ B
print(c)
print(C)
```

```python
## matrix operationns 
import numpy as np
print(np.linalg.det(A))
```
