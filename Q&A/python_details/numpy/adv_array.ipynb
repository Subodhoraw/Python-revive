## advanced array creation 

```python
#meshgrid (for plotting , calculation)
import numpy as np 
x = np.array([1,2,3])
y = np.array([4,5])
x , y = np.meshgrid(x,y)
print(x,y)
```

```python
# random sampling 
np.random.seed(42) #for reproducibility

#random choice from array
arr = np.array([1,2,3,4,5,6])
sample = np.random.choice(arr,size = 3,replace = False) #replace = false means it will never choose the same digit
print(sample)

#Random shuffle
print(np.random.shuffle(arr))
print(arr)

#random with specefic distribution 
normal = np.random.normal(loc = 0, scale = 1, size = 1000)
print(normal)

uniform = np.random.uniform(low = 0, high = 10,size = 1000)
print(uniform)
```

```python
arr = np.array([1,2,3,4,5])

#view (shares memory)
view = arr[:]
view[0] = 999
print(arr) # - original changed 

#copy(independent)
copy = arr.copy()
copy[0] = 111


#check if view or copy
print(view.base is arr)
print(copy.base is None)
```

```python
#medical image processing 
np.random.seed(42) 
image = np.random.randint(0,256,(100,100)) # 100x100 pixels

print("\n" + "="*60)
print(f"{'Medical image processing' :^60}")
print("="*60)

# image statistics
print(f"\nImage shape: {image.shape}")
print(f"Image size : {image.size} pixels")
print(f"Min Intensity: {np.min(image)}")
print(f"Max intensity: {np.max(image)}")
print(f"Mean intensity: {np.mean(image): .2f}")


#appply threshold(segment image)
threshold = 128
binary_image = image > threshold
print(f"\nPixels above thresold: {np.sum(binary_image)}")
print(f"Percentage : {np.mean(binary_image)* 100:.1f}%")

#histogram(intensity distribution)
hist, bins = np.histogram(image, bins = 10) # it means it will divide all values into 10 groups
print(f"\nIntensity distribution(10bins):")
for i in range(len(hist)):
    print(f"{bins[i]:.0f}-{bins[i+1]:.0f}:{hist[i]} pixels")

#apply smoothing
kernel_size = 3

smoothed = np.zeros_like(image,dtype = float)
```

Histogram Returns TWO Things

Actually:

counts, bin_edges = np.histogram(image, bins=10)


counts → number of pixels in each bin

bin_edges → the boundaries of the bins

Example:

counts = [200, 350, 900, 600, ...]
bin_edges = [0, 25, 50, 75, 100]


Meaning:

Between 0–25, you have 200 pixels
